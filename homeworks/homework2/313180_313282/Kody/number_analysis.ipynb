```python
from utils import get_aml_data
import matplotlib.pyplot as plt
```

```python
df, _ = get_aml_data()
df.head()
```

```python
largest_value = df.values.max()
smallest_value = df.values.min()

print(f"Largest Value: {largest_value}")
print(f"Smallest Value: {smallest_value}")
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df.values.flatten(), bins=100, edgecolor='black')
plt.title('Value Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
lower_bound = 400
upper_bound = 600

mask = (df.values.flatten() >= lower_bound) & (df.values.flatten() <= upper_bound)
percentage_in_range = (mask.sum() / len(mask)) * 100

print(f"Percentage of values in the range ({lower_bound}, {upper_bound}): {percentage_in_range:.2f}%")
```

```python
column_means = df.mean()
standard_deviation = df.std()

figure, (left, right) = plt.subplots(1, 2)
figure.set_size_inches(12, 6)

left.bar(df.columns, column_means.values)
left.set_title('Average of Each Column')
left.set_xlabel('Column')
left.set_ylabel('Average Value')

right.bar(df.columns, standard_deviation.values)
right.set_title('Standard Deviation of Each Column')
right.set_xlabel('Column')
right.set_ylabel('Standard Deviation')

plt.show()
```

```python
total_missing_values = df.isnull().sum().sum()
print("Total number of missing values in the DataFrame:", total_missing_values)
```
