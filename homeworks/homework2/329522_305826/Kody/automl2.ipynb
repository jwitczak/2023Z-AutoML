# Model autosklearn

```python
def get_data_and_scoring_function(data):
    X=data.iloc[:,:-1]
    y=data.iloc[:,-1]
    X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(
        X, y, random_state=42, stratify=y,
    )

    def scoring_function(estimator):
        pred = estimator.predict_proba(X_test)[:, 1]
        pred[pred>=0.5]=1
        pred[pred<0.5]=-1
        return sklearn.metrics.balanced_accuracy_score(y_test, pred)

    def train_scoring_function(estimator):
        pred = estimator.predict_proba(X_train)[:, 1]
        pred[pred>=0.5]=1
        pred[pred<0.5]=-1
        return sklearn.metrics.balanced_accuracy_score(y_train, pred)


    def get_test_data():
        return X_test, y_test

    return X_train, y_train, get_test_data, scoring_function, train_scoring_function

X_train_a, y_train_a, get_test_data, scoring_function, train_scoring_function = get_data_and_scoring_function(data)
```

resampling_strategy = "holdout"

```python
settings = {
  "time_left_for_this_task": 300,
  "seed": 2024,
  "n_jobs": 4,
  "metric":balanced_accuracy
}

resampling_strategy = "holdout"
```

```python
askl = AutoSklearnClassifier(
    **settings,
    resampling_strategy=resampling_strategy
)
askl.fit(X_train_a, y_train_a, dataset_name="symulacyjne")
```

```python
print(f"Auto-sklearn 1.0 | train = {train_scoring_function(askl)} | test = {scoring_function(askl)}")
print(f"Selected `resampling-strategy` = {askl.resampling_strategy}")
print(f"Selected `resampling-strategy-arguments` = {askl.resampling_strategy_arguments}")
```

```python
print(askl.sprint_statistics())

leaderboard = askl.leaderboard(sort_by="model_id", ensemble_only=True)
print(leaderboard)
```

```python
forest = RandomForestClassifier(random_state=2024)
forest.fit(X_train1, y_train)

feature_names = X_train1.columns
importances = forest.feature_importances_
forest_importances = pd.Series(importances, index=feature_names)
forest.fit(X_train1, y_train)
```

```python
plt.figure(figsize=(15,8))
plt.bar(forest_importances.index, forest_importances)
```

```python
print(forest_importances.sort_values(ascending=False).head(20))
selected_columns = forest_importances[forest_importances > 0.0025].index
print(selected_columns)
```

```python
X_train_sel = X_train1[selected_columns]
X_test_sel = X_test1[selected_columns]
```

```python
forest_sel = RandomForestClassifier(random_state=2023)
forest_sel.fit(X_train_sel, y_train)
```

```python
y_train_pred = forest_sel.predict(X_train_sel)
y_test_pred = forest_sel.predict(X_test_sel)
```

```python
def wyniki(y_train, y_train_pred, y_test, y_test_pred):
  df = pd.DataFrame({"zbiór": ['train', 'test'],
                    'balanced_accuracy': [balanced_accuracy_score(y_train, y_train_pred), balanced_accuracy_score(y_test, y_test_pred)],
                    #'accuracy': [accuracy_score(y_train, y_train_pred), accuracy_score(y_test, y_test_pred)],
                    "f1": [f1_score(y_train, y_train_pred), f1_score(y_test, y_test_pred)]})
  return df
wyniki(y_train, y_train_pred, y_test, y_test_pred)
```

Model jest mocno przeuczony.

```python
grid = {
    "n_estimators": [50, 100, 150, 200],
    "max_depth": [5,6,7,8,9,10],
    "min_samples_leaf": [50, 100],
    "max_features": [0.3, 0.6, 0.8, 1]
}


rs = RandomizedSearchCV(
    estimator = RandomForestClassifier(random_state=2023),
    param_distributions = grid,
    scoring = 'balanced_accuracy',
    cv = 5
)

search = rs.fit(X_train_sel, y_train.values.ravel())

search.best_params_
```

```python
search.best_score_
```

```python
rf = search.best_estimator_
y_train_pred = rf.predict(X_train_sel)
y_test_pred = rf.predict(X_test_sel)
wyniki(y_train, y_train_pred, y_test, y_test_pred)
```
