```python
from autosklearn.classification import AutoSklearnClassifier
from autosklearn.experimental.askl2 import AutoSklearn2Classifier
from autosklearn.metrics import roc_auc, balanced_accuracy

import pandas as pd
```

```python
X_train = pd.read_csv('https://raw.githubusercontent.com/kozaka93/2023Z-AutoML/main/homeworks/homework2/artificial_train.data',
                         sep=' ', header=None).iloc[:, :-1]
y_train = pd.read_csv("https://raw.githubusercontent.com/kozaka93/2023Z-AutoML/main/homeworks/homework2/artificial_train.labels", sep=' ', header=None).replace(-1, 0, inplace=False)
```

```python
X_val = pd.read_csv('https://raw.githubusercontent.com/kozaka93/2023Z-AutoML/main/homeworks/homework2/artificial_test.data',
                         sep=' ', header=None).iloc[:, :-1]
```

```python
X_val
```

```python
settings = {
  "time_left_for_this_task": 7200,
  "seed": 42,
  "metric": balanced_accuracy,
  "n_jobs": -1,
}

askl = AutoSklearn2Classifier(
    **settings
)
```

```python
askl.fit(X_train, y_train, dataset_name="artificial")
```

```python
askl.sprint_statistics()
```

```python
y_val = askl.predict_proba(X_val)
```

```python
pd.Series(y_val[:,1], name='313343_317028').to_csv("313343_317028_artifical_automl_prediction.txt", index=False)
```

```python
for model in askl.show_models().values():
  print(f'Model weight: {model["ensemble_weight"]}')
  print(repr(model["classifier"].choice))
  print(repr(model["data_preprocessor"].choice))
```
