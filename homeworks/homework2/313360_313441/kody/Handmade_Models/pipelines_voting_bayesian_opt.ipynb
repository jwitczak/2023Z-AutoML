```python
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import make_scorer
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from scipy.stats import randint as sp_randint
from scipy.stats import uniform
from skopt import BayesSearchCV
from sklearn.ensemble import VotingClassifier
from scipy.stats import randint as sp_randint
from sklearn.feature_selection import SelectFpr, SelectFdr
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
import numpy as np
#requires numpy==1.19.5 and pandas==1.3.4
```

```python
df_ultimate_test = pd.read_table("artificial_test.data", sep=" ", header=None)
df_ultimate_test = df_ultimate_test.drop(df_ultimate_test.columns[500], axis=1) # kolumna 500 do wyrzucenia
df_ultimate_test
```

```python
df_data_X = pd.read_table("artificial_train.data", sep=" ", header=None)
df_data_X = df_data_X.drop(df_data_X.columns[500], axis=1)
df_data_X
```

```python
df_data_Y = pd.read_table("artificial_train.labels", sep=" ", header=None)
df_data_Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_data_X, df_data_Y, test_size=0.33, random_state=4, stratify=df_data_Y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier


pipe = Pipeline([
    ('reduce_dim', 'passthrough'),
    ('clf', VotingClassifier(
            voting='soft',
            verbose=True,
            estimators=[
            ('knn1', KNeighborsClassifier()),
            ('knn2', KNeighborsClassifier()),
            ('knn3', KNeighborsClassifier()),
            ('rf1', RandomForestClassifier()),
            ('rf2', RandomForestClassifier()),
            ('rf3', RandomForestClassifier()),
            ('svc1', SVC(probability=True)),
            ('svc2', SVC(probability=True)),
            ('svc3', SVC(probability=True)),
            ('et1', ExtraTreesClassifier()),
            ('et2', ExtraTreesClassifier()),
            ('et3', ExtraTreesClassifier()),
            ('bc1', BaggingClassifier(KNeighborsClassifier(), random_state=4)),
            ('bc2', BaggingClassifier(KNeighborsClassifier(), random_state=4)),
            ('bc3', BaggingClassifier(KNeighborsClassifier(), random_state=4)),
            ('dt1', DecisionTreeClassifier()),
            ('dt2', DecisionTreeClassifier()),
            ('dt3', DecisionTreeClassifier()),
            ]
    ))
])
```

```python
model_distributions=[
    {
        'clf__knn1__n_neighbors': [3,4,5,6,7],
        'clf__knn2__n_neighbors': [3,4,5,6,7],
        'clf__knn3__n_neighbors': [3,4,5,6,7],
        'clf__rf1__n_estimators': list(range(100, 501)),
        'clf__rf2__n_estimators': list(range(100, 501)),
        'clf__rf3__n_estimators': list(range(100, 501)),
        'clf__rf1__max_depth': list(range(5, 13))+ [None],
        'clf__rf2__max_depth': list(range(5, 13))+ [None],
        'clf__rf3__max_depth': list(range(5, 13))+ [None],
        'clf__svc1__C': list(np.random.uniform(0.1, 10, size=10)),
        'clf__svc2__C': list(np.random.uniform(0.1, 10, size=10)),
        'clf__svc3__C': list(np.random.uniform(0.1, 10, size=10)),
        'clf__et1__n_estimators': list(range(100, 501)),
        'clf__et2__n_estimators': list(range(100, 501)),
        'clf__et3__n_estimators': list(range(100, 501)),
        'clf__et1__max_depth': list(range(5, 13))+ [None],
        'clf__et2__max_depth': list(range(5, 13))+ [None],
        'clf__et3__max_depth': list(range(5, 13))+ [None],

        'clf__bc1__n_estimators': list(range(9,200)),
        'clf__bc2__n_estimators': list(range(9,200)),
        'clf__bc3__n_estimators': list(range(9,200)),
        'clf__bc1__bootstrap': [True, False],
        'clf__bc2__bootstrap': [True, False],
        'clf__bc3__bootstrap': [True, False],

        'clf__dt1__criterion': ['gini', 'entropy', 'log_loss'],
        'clf__dt2__criterion': ['gini', 'entropy', 'log_loss'],
        'clf__dt3__criterion': ['gini', 'entropy', 'log_loss'],
        'clf__dt1__splitter': ['best', 'random'],
        'clf__dt2__splitter': ['best', 'random'],
        'clf__dt3__splitter': ['best', 'random'],
        'clf__dt1__max_features': ['sqrt', 'log2', None],
        'clf__dt2__max_features': [None, 'sqrt', 'log2'],
        'clf__dt3__max_features': [None, 'sqrt', 'log2'],
    }
]
```

```python
dimgrid=[
    {
        'reduce_dim': [PCA(svd_solver='full')],
        'reduce_dim__n_components': [4,5,6,7,8,499],
    },
    {
        'reduce_dim': [SelectFpr()],
    },
        {
        'reduce_dim': [SelectFdr()],
    },
]
balanced_accuracy_scorer = make_scorer(balanced_accuracy_score)
param_distributions=[]

for model in model_distributions:
    for dimred in dimgrid:
        param_distributions.append({**dimred, **model})
param_distributions
```

```python
bayes_search = BayesSearchCV(pipe, param_distributions, n_iter=400, scoring=balanced_accuracy_scorer, cv=4, n_jobs=-1, random_state=1, verbose=4, error_score='raise')
bayes_search.get_params(deep=True)
```

```python
bayes_search.fit(X=X_train, y=y_train)
```

```python
print(f"best trainig score {bayes_search.best_score_} ,score test {bayes_search.score(X_test, y_test)}")
```

```python
print("best params: %s" % str(bayes_search.best_params_))
```
