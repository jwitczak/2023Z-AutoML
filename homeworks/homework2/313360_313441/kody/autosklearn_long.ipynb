```python
import pandas as pd
df_ultimate_test = pd.read_table("artificial_test.data", sep=" ", header=None)
df_ultimate_test = df_ultimate_test.drop(df_ultimate_test.columns[500], axis=1) # kolumna 500 do wyrzucenia
df_ultimate_test
```

```python
df_data_Y = pd.read_table("artificial_train.labels", sep=" ", header=None)
df_data_Y.columns = ["label"]
df_data_Y
```

```python
df_data_X = pd.read_table("artificial_train.data", sep=" ", header=None)
df_data_X = df_data_X.drop(df_data_X.columns[500], axis=1)
df_data_X
```

```python
import autosklearn.classification
import autosklearn.metrics
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_data_X, df_data_Y, test_size=0.33, random_state=4, stratify=df_data_Y)
```

```python
automl = autosklearn.classification.AutoSklearnClassifier(metric=autosklearn.metrics.balanced_accuracy,n_jobs=-1, time_left_for_this_task=28800)
automl.fit(X_train, y_train)
```

```python
pred = automl.predict(X_test)
tpred = automl.predict(X_train)
```

```python
print(f"Auto-sklearn | train = {autosklearn.metrics.balanced_accuracy(y_train, tpred)} | test = {autosklearn.metrics.balanced_accuracy(y_test, pred)}")
print(f"Selected `resampling-strategy` = {automl.resampling_strategy}")
print(f"Selected `resampling-strategy-arguments` = {automl.resampling_strategy_arguments}")

print(automl.sprint_statistics())

leaderboard = automl.leaderboard(sort_by="model_id", ensemble_only=True)
print(leaderboard)
```

```python
from autosklearn.experimental.askl2 import AutoSklearn2Classifier
```

```python
automl2 = AutoSklearn2Classifier(metric=autosklearn.metrics.balanced_accuracy, n_jobs=-1, time_left_for_this_task=28800)
automl2.fit(X_train, y_train)
```

```python
pred = automl2.predict(X_test)
tpred = automl2.predict(X_train)
```

```python
print(f"Auto-sklearn 2.0 | train = {autosklearn.metrics.balanced_accuracy(y_train, tpred)} | test = {autosklearn.metrics.balanced_accuracy(y_test, pred)}")
print(f"Selected `resampling-strategy` = {automl2.resampling_strategy}")
print(f"Selected `resampling-strategy-arguments` = {automl2.resampling_strategy_arguments}")

print(automl2.sprint_statistics())

leaderboard = automl2.leaderboard(sort_by="model_id", ensemble_only=True)
print(leaderboard)
```
