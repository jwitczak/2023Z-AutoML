```python
import pandas as pd

data_train = pd.read_table('data/artificial_train.data', sep=" ", header=None)
y_train = pd.read_table('data/artificial_train.labels', sep=" ", header=None)
```

```python
unseen_data = pd.read_table('data/artificial_test.data', sep=" ", header=None)
```

```python
unseen_data
```

```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.model_selection import train_test_split

data_train.drop(columns=[64, 442, 28, 105, 153, 241, 318, 281, 453, 48, 500], inplace=True)

data_train.columns = data_train.columns.astype(str)

X_train, X_test, y_train, y_test = train_test_split(data_train, y_train, test_size = 0.2, random_state=123)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(random_state=123).fit(X_train, y_train)

forest_coefs = pd.Series(forest.feature_importances_, index=forest.feature_names_in_)

forest_coefs.sort_values(ascending=False, inplace=True)

# plt.figure(figsize=(15, 8))
# forest_coefs[:60].plot.barh()
```

```python
y_train[0]
```

```python
X_train['y'] = y_train[0]
```

```python
X_train
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor

predictor = TabularPredictor('y', eval_metric='balanced_accuracy').fit(X_train)
```

```python
predictor.leaderboard()
```

```python
preds = predictor.predict(X_test)

balanced_accuracy_score(y_test, preds)
```
