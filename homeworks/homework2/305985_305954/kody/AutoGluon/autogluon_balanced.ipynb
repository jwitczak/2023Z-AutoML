# Model using AutoGluon framework

For this part we created a model using AutoGluon `TabularPredictor` with `balanced accuracy` as its evaluation function.

## Loading data and setting up working environment

```python
#!pip install autogluon
```

```python
import pandas as pd
from sklearn.metrics import balanced_accuracy_score
from autogluon.tabular import TabularPredictor
```

```python
a_pred = pd.read_csv("https://raw.githubusercontent.com/kozaka93/2023Z-AutoML/main/homeworks/homework2/example_artifical_prediction.txt")
a_test = pd.read_csv("https://raw.githubusercontent.com/kozaka93/2023Z-AutoML/main/homeworks/homework2/artificial_test.data", header=None, sep=" ")
a_train = pd.read_csv("https://raw.githubusercontent.com/kozaka93/2023Z-AutoML/main/homeworks/homework2/artificial_train.data", header=None, sep=" ")
a_train_labels = pd.read_csv("https://raw.githubusercontent.com/kozaka93/2023Z-AutoML/main/homeworks/homework2/artificial_train.labels",header=None)
```

```python
a_train = a_train.drop(columns=500, axis=1)
a_test = a_test.drop(columns=500, axis=1)
```

```python
a_train_labels = a_train_labels.rename({0:"y"}, axis=1)
```

```python
complete_df = pd.concat([a_train_labels, a_train], axis=1)
```

## Model

```python
label= "y"
```

```python
m_gluon_balanced = TabularPredictor(label=label, eval_metric = "balanced_accuracy")
```

```python
m_gluon_balanced.fit(complete_df)
```
