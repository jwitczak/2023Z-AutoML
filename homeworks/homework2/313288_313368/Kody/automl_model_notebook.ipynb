## Wczytanie danych jako DataFrames

Dane zostały pobrane i połączone w jedną tabelę

```python
import pandas as pd

train_X  = pd.read_csv("Data/artificial_train.data", sep=" ", header=None)

# Dropping column, which contains only NaN values
train_X.drop(train_X.columns[500], axis=1, inplace = True)

train_Y = pd.read_csv("Data/artificial_train.labels", sep=" ", header=None)
train_Y.rename(columns={train_Y.columns[0]: 'Y'},inplace=True)

data = pd.concat([train_X, train_Y], axis=1)
```

```python
data
```

## Podział danych na zbiór treningowy i testowy

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.2, random_state=42, shuffle=False)
```

## Trenowanie modelu

```python
from autogluon.tabular import TabularDataset, TabularPredictor

label = 'Y'
save_path = 'autoGluon-hw2-model'  # specifies folder to store trained models

predictor = TabularPredictor(
    label=label,
    path=save_path,
    eval_metric='balanced_accuracy',
    log_to_file=True
)

predictor = predictor.fit(train_data, presets="best_quality")
```

## Sprawdzenie jakości uzyskanego modelu

```python
y_pred = predictor.predict(test_data.drop(columns=[label]))
y_pred.head()
```

```python
predictor.evaluate(test_data, silent=True)
```

```python
predictor.leaderboard(test_data)
```

## Zapis wyników do pliku

```python
test_X  = pd.read_csv("Data/artificial_train.data", sep=" ", header=None)
test_X.drop(test_X.columns[500], axis=1, inplace = True)
```
